```python
import zipfile
extract_path = "/content"
with zipfile.ZipFile("/content/sw.zip", 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"파일이 {extract_path} 경로에 성공적으로 압축 해제되었습니다.")

# /content/Swimming_and_Drowning.zip
```

```python
import yaml

data_yaml_path = '/content/data.yaml'
with open(data_yaml_path, 'r') as file:
    data = yaml.safe_load(file)

data['train'] = '/content/train/images'
data['val'] = '/content/valid/images'
data['test'] = '/content/test/images'

with open(data_yaml_path, 'w') as file:
    yaml.safe_dump(data, file)
```
